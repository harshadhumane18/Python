```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

## Categorical scatter Plots:
### (Bivariate analysis)

```python
# -Categorical Scatter Plot (one column should be categoricak and another should be nummerical)
#     • Stripplot
#     • Swarmplot
# -Categorical Distribution Plots
#     • Boxplot
#     • Violinplot
# -Categorical Estimate Plot -> for central tendency
#     • Barplot
#     • Pointplot
#     • Countplot
```

## -Strip plot

```python
tips.head()
```

```python
# axes level function
sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
# to make the scatter points into line, use jitter parameter
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
```

```python
# figure level function for categorical plot using catplot
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='strip')
```

## - Swarm plot

```python
# using axes plot (quite better visual distribution, on small dataset use)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
# using figure plot 
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='swarm')
```

## ~ Categorical distribution Plots:

```python
# univariate analysis
# -boxplot
# -voiline plot
```

### - Boxplot(based on 5 no. summary)

```python
# axes level function
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
# figure level function
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='box')
```

```python
# single boxplot -> numerical col
sns.boxplot(data=tips,y='total_bill')
```

## - Violinplot = (Boxplot + KDEplot)

```python
# axes level function
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
# figure level function
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='violin')
```

## - Categorical Estimate Plot -> for central tendency

```python
# estimate values(aggregated value): estimated mean,count,std deviation 
# • Barplot
# • Pointplot
# • Countplot
```

### - Bar plot

```python
# can cal. aggregated value of one categorical column with resp to one numerical column
# there's paramter called estimator(bydefault=mean) you change that to other estimation too
```

```python
# axes level function
sns.barplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
# figure level function
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='bar')
```

## - Countplot

```python
# A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a
# second variable. This is similar to a histogram over a categorical, rather than quantitative, variable
```

```python
sns.countplot(data=tips,x='sex',hue='day')
```

```python
# facetplot using catplot
# Facetplot --> used to draw separate multiple plots for multiple category for categorical column.
# facet plot -> work with nly figure level function 
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',col='smoker',row='day',kind='bar')
```
