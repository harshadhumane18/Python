```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Figure level function : relplot, displot,catplot
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## ~Scatterplot

```python
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex',size='size',style='time')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip',hue='sex',size='size',style='time',kind='scatter')
```

## ~Line plot

```python
import plotly.express as px
```

```python
gap = px.data.gapminder()
gap.head()
```

```python
temp_df = gap[gap['country']=='India']
temp_df.head()
```

```python
sns.lineplot(data=temp_df, x='year',y='lifeExp')
```

```python
temp_df1 = gap[gap['country'].isin(['India','Germany','Brazil'])]
temp_df1.head()
```

```python
# figure function --> kind = 'line'
sns.relplot(data=temp_df1, x='year',y='lifeExp',hue='country',kind='line',style='continent')
```

## ~Facetplot

```python
# Facetplot --> used to draw separate multiple plots for multiple category for categorical column.
# facet plot -> figure level function -> work with relplot only
# it will not work with scatterplot and lineplot (axes level function)
tips
```

```python
# column wise plotting
sns.relplot(data=tips, x='total_bill', y='tip', col='sex')
```

```python
# row wise plotting
sns.relplot(data=tips, x='total_bill', y='tip', row='sex')
```

```python
# both row and column wise plotting
sns.relplot(data=tips, x='total_bill', y='tip', col='smoker', row='sex')
```

```python
# if there are lots of category in categorical column. Soln-->> col_wrap-by which you can limit the graph to appear in row
# example:
temp_df
```

```python
sns.relplot(data=gap, x='lifeExp',y='gdpPercap',col='year', col_wrap=3, kind='scatter')
```

## ~Distribution plot

```python
# Distribution Plots:
# - used for univariate analysis
# - used to find out the distribution
# - ange of the observation
# - entral Tendency
# - is the data bimodal?
# - Are there outliers?
```

```python
# figure level - displot
# axes level -> histplot - kdeplot - rugplot
```

## -Histogram

```python
tips
```

```python
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
# there's hue parameter to distinguish between category
# element parameter to make clear vision about category
sns.displot(data=tips,x='tip',kind='hist',hue='sex',element='step')
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data=tips,x='tip',kind='hist',col='sex')
```

## - KDE plot

```python
#kdeplot
''' Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel,
   producing a continuous density estimate. rather than couting its eastimate the probability of specific point in dataset
'''
```

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
# hue -> fill
sns.displot(data=tips,x='total_bill',hue='sex',fill=True,kind='kde')
```

## -Rug plot

```python
# Rugplot
# Plot marginal distributions by drawing ticks along the x and y axes.
# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
```

```python
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Note: generally the distribution plot we saw above were of univariate analysis
# to use distribution plot for bivariate analysis used below format
```

```python
# bivariate histogram
sns.histplot(data=tips,x='total_bill',y='tip')
```

```python
# bivariate kdeplolot
sns.kdeplot(data=tips,x='total_bill',y='tip')
```

## Heatmap

```python
gap
```

```python
# to convert the long dataset to wide dataset 
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')
temp_df
```

```python
# heatmap is axes level
plt.figure(figsize=(20,20)) #--> To increase the size of figure use matplot lib function 
sns.heatmap(temp_df)
```

```python
# using heatmap ploting only those countries which are from europe continent
europe = gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
europe
```

```python
sns.heatmap(europe)
```

```python
# other customization parameter for heatmap: annot(to get value resp of colour),linewidth (border to boxes),cmap(summer,greens,etc)
plt.figure(figsize=(20,20))
sns.heatmap(europe,annot=True,linewidth=0.5,cmap='summer')
```
