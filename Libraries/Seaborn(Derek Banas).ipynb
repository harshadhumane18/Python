```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
print(sns.get_dataset_names())
```

```python
df = sns.load_dataset('car_crashes')
df
```

# Distribution plot

```python
# Distribution plots(also a histogram),plot a univariate distribution of data.
# Univariate describe a type of data which consists of observations on only a single characteristic or attribute.
# A simple example of univariate data would be the salaries of workers in industry.
# kde = ('~') kernal density estmation (estimation : judgement,opinion )
# shift + Tab : will provide the no. of argument we can provide

sns.distplot(df['not_distracted'])

plt.show()
```

# Jointplot

```python
''' The function jointplot() in Seaborn library creates a scatter plot with two histograms at the top 
   and right margins of the graph by default. Where, univariate and bivariate plot, plots together in one figure
   bivariate : explores the relationship between two variables and also describes the strength of their relationship. '''

# for regression line
sns.jointplot(x='speeding', y='not_distracted', data=df, kind='reg')

plt.show()
```

```python
# for kde plot
sns.jointplot(x='speeding', y='not_distracted', data=df, kind='kde')

plt.show()
```

```python
# for hexagon distribution plot
sns.jointplot(x='speeding', y='not_distracted', data=df, kind='hex')

plt.show()
```

# Styling to the plot (style, font size, font style)

```python
# to change the style of the plot : darkgrid,dark,whitegrid,white,ticks
sns.set_style('ticks')

# to change the size of figure
plt.figure(figsize=(8,4))

# to change the font and size of the lebels:paper,talk,poster
sns.set_context('paper', font_scale = 1.2) 

sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')

# to hide the axis line on plot
# sns.despine(left=True, bottom=True)

plt.show()
```

# Categorical plot: Bar plot

```python
# Categorical plot : will look for distribution of categorical data with reference of one of the numreical data

# you can also aggregate the data on the basis of statistics term using argument: estimator = np.median,cov,mean,std
sns.barplot(x='sex', y='total_bill', data=df1, estimator = np.std)
```

# Count plot

```python
# it just count the occurances the data given
sns.countplot(x='sex', data=df1)
```

# box plot

```python
# Quartile : measures the spread of values above and below the mean by dividing the distribution into four groups.
# box plot :used to show distributions of numeric data values, especially when you want to compare them between multiple groups.

sns.boxplot(x='day', y='total_bill', data=df1, hue = 'sex')
```

# violin plot

```python
# combination of box plot and kde plot
# box plot corresponds to data points and violin plot used kde estimation of data point
# splits=True : shows camparision between categories

sns.violinplot(x='day', y='total_bill', data=df1, hue = 'sex')
```

# Strip plot 

```python
# A density plot. 
'''A strip plot is a single-axis scatter plot that is used to visualise the distribution of many individual 
   one-dimensional values. The values are plotted as dots along one unique axis, and the dots with the same 
   value can overlap.'''
# jitter : used to spred the data little bit to look little convenient
# dodge : separate the data based on hue or category like male and female

# before dodge
sns.stripplot(x='day', y='total_bill', data=df1, jitter=True, hue = 'sex' )
```

```python
sns.stripplot(x='day', y='total_bill', data=df1, jitter=True, dodge=True, hue = 'sex' )
```

# Swarm plot

```python
'''swarmplot: is probably similar to stripplot, only the points are adjusted so it wont get overlap 
   to each other as it helps to represent the better representation of the distribution of values.'''

sns.swarmplot(x='day', y='total_bill', data=df1)
```

```python
'''Swarm plot : can be drawn on its own, but it is also a good complement to a box or violin plot in cases 
   where you want to show all observations along with some representation of the underlying distribution.'''

sns.violinplot(x='day', y='total_bill', data=df1)
sns.swarmplot(x='day', y='total_bill', data=df1, color = 'white')
```

# Palatte

```python
# Palatte: color map, lots of in built are available in seaborn
# to chance the style of the plot : darkgrid,dark,whitegrid,white,ticks
sns.set_style('whitegrid')

# to change the size of figure
plt.figure(figsize=(8,4))

# to change the font and size of the plot/lebels:paper,talk,poster
sns.set_context('paper') 

sns.stripplot(x='day', y='total_bill', data=df1, jitter=True, hue = 'sex', palette='summer' )

# location for legend
# upperleft2:, lowerleft3:, upperright:1, lowerright:4
plt.legend(loc=2)
```

# Heatmap

```python
'''Heatmap: Show correlation between two or more attributes and is very useful in visualizing the concentration of value
   between two dimensions of a matrix. This helps in finding patterns and gives a perspective of depth.'''

flights = sns.load_dataset('flights')
# there should be correlation betweeen the attributes of data frame and there are two ways
# 1st by pivot table:(preferable)
flights = flights.pivot_table(index = 'month', columns = 'year', values ='passengers')
flights
```

```python
sns.heatmap(flights, cmap='Blues')
```

```python
# 2nd metthod to coorelate 
tips_mx = df1.corr()
tips_mx
sns.heatmap(flights_mx, annot=True, cmap='Blues')
```

```python
tips = df1.pivot_table(index = 'day', columns = 'sex', values ='tip')
sns.heatmap(tips, annot=True, cmap='Blues')
```
