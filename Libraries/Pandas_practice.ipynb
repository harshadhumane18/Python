```python
# IMBD data set
df = pd.read_csv("imdb_data.csv")
df.head()
```

```python
eng_movies = df[df["original_language"] == "en"]
eng_movies.head()
```

```python
# movies which have runtime more than 100min
runtym = df[df['runtime']>= 100]
runtym
```

```python
# english movies which are above the runtime of 100min
cond1 = df[(df["original_language"] == "en") & (df["runtime"]>120)]
cond1
```

```python
# to create new column 
df['runtime_morethan_120'] = df['runtime']>120
df
```

```python
# new column to know the profit generated by movie
df["Profit_generated"] = df['revenue']-df['budget']
df.head(10)
```

```python
# to remove particular column or rom
df.drop('runtime_morethan_120', axis=1)
# we cam also provide list of column to drop by providing list of column
```

```python
# if want to set particular column to be index 
df.set_index('id')
```

```python
# apply() method applies a function to every row in pandas dataframe

def uppercase_title(column):
    return column.upper()
```

```python
df['Uppercase_title'] = df['original_title'].apply(uppercase_title)
df
```
